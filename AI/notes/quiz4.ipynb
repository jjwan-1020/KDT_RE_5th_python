### 문제 7
- 텐서 [[1, 2, 3], [4, 5, 6]]를 2개의 조각으로 나눠서 출력해보세요. (행 기준).

````python
# 여기에 코드를 작성합니다
t = torch.tensor([[1,2,3],[4,5,6]])
t1, t2 = torch.chunk(t, 2, dim=0)
print(t1)
print(t2)
````

### 문제9
- 텐서 [[1, 2], [3, 4], [5, 6]]에서:
- dim=1 기준으로 크기 1씩 나누세요.
- 각각의 조각을 새로운 축에 쌓은 후 출력해보세요.

````python
# 여기에 코드를 작성합니다
t = torch.tensor([[1, 2], 
                  [3, 4], 
                  [5, 6]])

# 1. 열 기준으로 크기 1씩 나누기
tp = torch.split(t, 1, dim=1)
print(tp)

# 2. 새로운 축에 쌓기
newt = torch.stack(tp, dim=0)
print(newt)
````

### 문제10
- 다음 텐서를 활용해서 아래 문제를 해결해보세요
  ```
  a = torch.tensor([[1, 2], [3, 4]])
  b = torch.tensor([[5, 6], [7, 8]])
  ```

1. 두 텐서를 열 기준으로 연결하고 결과를 출력하세요.
2. 연결된 텐서를 크기 `(4, 2)`로 변환하세요.
3. 변환된 텐서를 2개의 조각으로 나누고, 각 조각을 0번 축으로 스택하여 출력하세요.


````python
# 여기에 코드를 작성합니다
a = torch.tensor([[1, 2], 
                  [3, 4]])
b = torch.tensor([[5, 6], 
                  [7, 8]])
# 1. 열 기준으로 병합
tc = torch.cat([a,b], dim=1)
print("1번.:", tc)

# 2. 크기 (4, 2)로 변환
ts = tc.view(4,2)
print("2번:", ts)

# 3. 2개의 조각으로 나누고
tch = torch.chunk(ts,2,dim=0)
print("3번:", tch)

# 4. 0번 축으로 스택
tl = torch.stack(tch, dim=0)
print("4번:", tl)
````

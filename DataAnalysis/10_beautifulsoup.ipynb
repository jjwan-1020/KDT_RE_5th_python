```python
# ë°ì´í„° ì„ íƒ
# find() - ì²« ë²ˆì§¸ ë§¤ì¹­ ìš”ì†Œ ì„ íƒ
# 1) íƒœê·¸ë¥¼ ê¸°ì¤€ìœ¼ë¡œ íƒìƒ‰
title_tag = soup.find("h1")
print(title_tag)
print(title_tag.text)
print(title_tag.get_text())
```

```python
# find() = ì†ì„± ì¡°ê±´ìœ¼ë¡œ ê²€ìƒ‰ ê°€ëŠ¥
result = soup.find("h1", class_="sub_title")
print(result.text)
```

```python
# find_all = ëª¨ë“  ë§¤ì¹­ëœ ìš”ì†Œë¥¼ ì„ íƒ
result = soup.find_all("h1")
print(result)

for i in result:
    print(i.text)
```

```python


url = "https://news.naver.com/section/102"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")

articles = soup.select("div.sa_item_flex a.sa_text_title")

for a in articles[:15]:
    title = a.get_text(strip=True)
    href = a.get("href")

  
    if href.startswith("/"):
        link = "https://news.naver.com" + href
    else:
        link = href

    print(f"- {title}\n,{link}\n")

```

```python

url = "https://news.naver.com/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "lxml")

articles = soup.select("strong.cnf_news_title")

for idx, articles in enumerate(articles):
    print(f"{idx+1}. {articles.text.strip()}")
    link = articles.find_parent('a')['href']
    print(f"{link}\n")
```

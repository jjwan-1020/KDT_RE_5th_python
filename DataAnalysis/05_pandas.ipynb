### 조건 필터링

```python
data = {
    '이름': ['홍길동', '김철수', '이영희', '박민수'],
    '점수': [90, 70 , 95, 95],
    '반' : [2,1,1,2]
}
df = pd.DataFrame(data)
df
```

```python
mask = df['점수'] >= 80
filtered = df[mask]
filtered
```

```python
# 여러 조건 결함
# and : &
# or : |
#  not : ~

# & 예시
df[(df["반"]== 2) & (df["점수"] >= 90)]

```

```python
#  not
df[~(df["점수"] >= 80)]
```

```python
# isin()
# 여러 값 중 하나에 해당하는지 여부를 판단할 때 사용

# ex. 이름이 홍길동 이거나 박민수 이거나 한 경우
# mask = (df["이름"] == "홍길동") | (df["이름"] == "박민수")
mask = df["이름"].isin(["홍길동", "박민수"])
df[mask]

mask2 = ~df["이름"].isin(["홍길동", "박민수"])
df[mask2]
```

```python
# reset_index
# 조건 필터링, 행 삭제 등을 통해 인덱스가 변경됐을 시
# reset_index()를 사용 -> 인덱스가 초기화

mask = (df["반"] == 2) | (df["점수"] >= 90)
df2 = df[mask]
print(df2)

# 인덱스 리셋
df2 = df2.reset_index()
print(df2)

# 기존 인덱스 삭제
df2 = df2.reset_index(drop=True)
print(df2)
```

### 정렬
- 값 기준 정렬: sort_values
- 인덱스 기준 정렬: sort_index

```python
data = {
    '이름': ['홍길동', '김철수', '이영희', '박민수'],
    '점수': [90, 70 , 85, 95],
    '반' : [2,1,1,2]
}
df = pd.DataFrame(data)
df
```

```python
# 오름차순 정렬
df2 = df.sort_values('점수').reset_index(drop=True)
df2
```

```python
# 내림차순 정렬
# ascending = False
df3 = df.sort_values('점수', ascending=False).reset_index(drop=True) 
df3
```

```python
# 여러 기준으로 정렬
df4 = df.sort_values(['반','점수'], ascending=[True, False]).reset_index(drop=True)
df4
```

```python
# 원본 변경
df.sort_values(['반','점수'], ascending=[True, False], inplace=True)
df2 = df.reset_index(drop=True)
df2
```

```python
# 인덱스 기준 정렬: sort_index()
df_shuffled = df.sample(frac=1, random_state=42)
df_shuffled

# 행 인덱스 기준으로 정렬
df5 = df_shuffled.sort_index()
df5

# 열 이름 기준으로 정렬 (알파벳순)
df6 = df.sort_index(axis=1)
df6
```

### groupby

```python
df = pd.DataFrame({
    'team': ['A','A','B','B','B','C'],
    'name': ['Kim','Lee','Park','Choi','Jung','Han'],
    'score': [90,85,80,70,95,88]
})
df

# 단일 컬럼 기준 그룹화
grouped = df.groupby('team')

# 집계함수 적용
result_sum = grouped['score'].sum()
print(result_sum)

# 평균
result_mean = grouped['score'].mean()
print(result_mean)

# 개수
result_count = grouped['score'].count()
print(result_count)
```

```python
df2 = pd.DataFrame({
    'team': ['A','A','B','B','B','C'],
    'gender': ['M','F','F','M','M','F'],
    'score': [90,85,80,70,95,88],
    'age' : [21, 22, 23, 25, 20, 27]
})
df2
```

```python
# 여러 컬럼 기준 그룹화
result = df2.groupby(['team','gender'])['score'].mean()
result
```

```python
# as_index=False 옵션
# 그룹라벨이 인덱스로 설정됨
result = df2.groupby('team', as_index=False)['score'].sum()
result
```

```python
# 여러 집계 함수를 한번에 적용: agg()
result = df2.groupby('team', as_index=False)['score'].agg(['sum','mean','count'])
result
```

```python
# 그룹별로 여러 컬럼에 다른 집계 함수 적용
result = df2.groupby('team').agg({
    'score': 'mean',
    'age': 'max'
})
result
```

```python
# 실습 4.
# 1.
df = pd.DataFrame({
    'grade': [1,2,1,2,1,3],
    'name': ['Kim','Lee','Park','Choi','Jung','Han'],
    'kor': [85,78,90,92,80,75]
})
an1 = df.groupby('grade')['kor'].mean()
print(an1)
```

```python
# 2.
df = pd.DataFrame({
    'class': [1,1,1,2,2,2],
    'subject': ['Math','Math','Eng','Math','Eng','Eng'],
    'score': [80,90,85,70,95,90]
})
an2 = df.groupby(['class','subject']).agg(
    count = ('score', 'count'),
    avg = ('score', 'mean')      
)
print(an2)
```

```python
# 3.
df = pd.DataFrame({
    'region': ['Seoul','Seoul','Busan','Busan','Daegu'],
    'seller': ['A','B','A','B','A'],
    'sales': [100,200,150,120,130]
})
an3 = df.groupby(['region','seller']).agg(
    t_sales = ('sales','sum'),
    m_sales = ('sales','max')
)
print(an3)
```

```python
# 4.
df = pd.DataFrame({
    'team': ['A','A','B','B','A','B'],
    'position': ['FW','DF','FW','DF','DF','FW'],
    'score': [3,2,None,1,4,2]
})
an4 = df.groupby(['team','position'])['score'].mean()
print(an4)
```

```python
# 5.
df = pd.DataFrame({
    'dept': ['HR','HR','IT','IT','Sales','Sales'],
    'gender': ['M','F','F','M','F','F'],
    'salary': [3500,3200,4000,4200,3000,3100]
})
an5 = df.groupby(['dept','gender']).agg(
    count = ('salary','count'),
    t_salary = ('salary','sum')
)
print(an5)
```

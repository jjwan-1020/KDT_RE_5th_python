```python
#  스타일 설정
sns.set_style("ticks") # 스타일 테마 설정
sns.set_context("notebook") # 문맥에 따라 스타일 크기 조정
sns.set_palette("pastel") # 스타일 색상
plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕이 설치되어 있는경우
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호(-)깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot()
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x="day", hue="sex", palette=palette)
plt.title("요일별 방문자수")
plt.grid(linestyle=":")
plt.show()
```

```python
# barplot
palette= sns.color_palette("colorblind")
sns.barplot(data=tips, x="day", y="tip", palette=palette)
plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# displot()
# 히스토그램 + 커널 밀도 (kde)
# kde = True 로 설정하면 kde 그래프를 그려준다
sns.displot(data=tips, x="total_bill", bins=20)
sns.displot(data=tips, x="total_bill", kde=True,  bins=20)
plt.title("Total bill")
plt.show()
```

```python
# scatterplot()
# 산점도

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
# countplot() 연습
palette = ("#4283E4","#e44514","#09d82c")
titanic = sns.load_dataset("titanic")
sns.countplot(data= titanic, x="class", hue="who", palette=palette)
plt.title("객실 등급별 탑승자유형")
plt.show()
```

```python
# displot() 연습
sns.displot(data=titanic, x="age", kde=True, bins=20)
plt.title("탑승자 나이 분포")
plt.show()
```

```python
# scatterplot() 연습

titanic = sns.load_dataset("titanic")
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived")
plt.title("나이에 따른 요금 분포 (생존 여부 구분)")
plt.xlabel("나이")
plt.ylabel("요금")
plt.show()
```

```python
# countplot() 연습
palette = ("#4283E4","#e44514","#09d82c")
titanic = sns.load_dataset("titanic")
sns.countplot(data= titanic, x="class", hue="who", palette=palette)
plt.title("객실 등급별 탑승자유형")
plt.show()

# barplot() 연습
palette=sns.color_palette(palette)
titanic = sns.load_dataset("titanic")
sns.barplot(data=titanic, x="class",y="survived", hue="who", palette=palette)
plt.title("등급별 탑승자 유형 분포")
plt.show()
print()

# displot() 연습
sns.displot(data=titanic, x="age", kde=True, bins=20)
plt.title("탑승자 나이 분포")
plt.show()

# scatterplot() 연습
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived")
plt.title("나이에 따른 요금 분포 (생존 여부 구분)")
plt.xlabel("나이")
plt.ylabel("요금")
plt.show()
```

```python
# heatmap()
# 상관계수행렬의 그래프
penguins = sns.load_dataset("penguins")

#상관계수행렬
# - 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# - heatmap에서 색으로 표시되는 숫자가 상관계수
# 1: 양의 상관관계 (하나가 증가하면 다른것도 같은 비율로 증가)
# ex) 키와 몸무게
# -1: 음의 상관관계 (하나가 증가하면 다른 변수는 감소)
# ex) 공부시간과 실수 개수

corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="inferno")
plt.show()
```

```python
# jointplot()
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="kde")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="hist")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="reg")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="hex")
plt.show()
```

```python
# 실습
numeric_cols = penguins.select_dtypes(include=np.number)
corr = numeric_cols.corr()
sns.heatmap(corr, annot=True, cmap="inferno", fmt=".2f")
plt.show()
```

```python
sns.jointplot(data=penguins,
              x='bill_length_mm',
              y='bill_depth_mm',
              hue='species',
              kind='scatter')
plt.suptitle("종별 부리 길이와 부리 깊이의 관계", y=1.02)
plt.show()
```

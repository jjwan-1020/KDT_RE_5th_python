## 09. Folium
- 지도 시각화 라이브러리


```python
import folium
```

```python
# 기본 지도 생성
m = folium.Map(location=[37.61112366688166, 126.9172029846246], zoom_start=18)
# 기본 마커 생성
folium.Marker([37.61112366688166, 126.9172029846246],
              popup="Subway",
              tooltip="구산역",
              icon=folium.Icon(color="red", icon="fa-solid fa-bus", prefix="fa-solid")).add_to(m)
m
```

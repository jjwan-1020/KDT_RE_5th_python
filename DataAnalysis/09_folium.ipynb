## 09. Folium
- 지도 시각화 라이브러리


```python
import folium
```

```python
# 기본 지도 생성
m = folium.Map(location=[37.61112366688166, 126.9172029846246], zoom_start=18)
# 기본 마커 생성
folium.Marker([37.61112366688166, 126.9172029846246],
              popup="Subway",
              tooltip="구산역",
              icon=folium.Icon(color="red", icon="fa-solid fa-bus", prefix="fa-solid")).add_to(m)
m

#원형 마커
folium.CircleMarker(
    [37.6103, 126.9128],
    radius=100,
    color="#adcdff",
    fill_color="#1c73ff",
    popup="CircleMarker",
    tooltip="tooltip"
).add_to(m)

# 클릭한 곳에 마커 추가
m.add_child(folium.ClickForMarker(popup='내가 클릭한곳'))

# 클릭한 곳의 위도, 경도 표시
m.add_child(folium.LatLngPopup())
m


```

```python
df = pd.read_csv("./dataset/택시승차대 현황.csv", encoding="cp949")

taxi = df.loc[:, ["위치명", "위도", "경도"]]
latitde = taxi["위도"].mean()
longitude= taxi["경도"].mean()
m = folium.Map([latitde, longitude], zoom_start=13)

for idx, row in taxi.iterrows():
    folium.Marker(
        [row["위도"], row["경도"]],
        tooltip=row["위치명"],
        icon=folium.Icon(color="orange", icon="fa-solid fa-taxi", prefix="fa-solid")
    ).add_to(m)

m
```
